```python
import pandas as pd

# Load the Excel file
file_path = "Experimental analysis.xlsx"  # Update the path if needed
df = pd.read_excel(r"C:\Users\user\Desktop\Experimental analysis.xlsx")

# Clean column names
df.columns = df.columns.str.strip().str.replace(r'\s+', ' ', regex=True)

# Gender distribution
print("Gender Distribution:")
print(df['GENDER'].value_counts(), "\n")

# Age statistics
print("Age Statistics:")
print(df['AGE'].describe(), "\n")

# Financial problems analysis
j1_cols = [col for col in df.columns if col.startswith('J1__')]
j1_counts = df[j1_cols].apply(lambda col: (col == 'Yes').sum()).sort_values(ascending=False)

print("Top 5 Reported Financial Problems:")
print(j1_counts.head(5), "\n")

# Digital access barriers analysis
k2_cols = [col for col in df.columns if col.startswith('K2__')]
k2_counts = df[k2_cols].apply(lambda col: (col == 'Yes').sum()).sort_values(ascending=False)

print("Top 5 Digital Access Barriers:")
print(k2_counts.head(5), "\n")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
#file_path = "Experimental analysis.xlsx"
#df = pd.read_excel(file_path, sheet_name="Sheet1")

# Clean column names
df.columns = df.columns.str.strip().str.replace(r'\s+', ' ', regex=True)

# Set visual style
sns.set(style="whitegrid")

# --- Gender Distribution Plot ---
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='GENDER', palette='Set2')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# --- Age Distribution Histogram ---
plt.figure(figsize=(6, 4))
sns.histplot(df['AGE'], bins=10, kde=True, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# --- Top Financial Problems (J1__) ---
j1_cols = [col for col in df.columns if col.startswith('J1__')]
j1_counts = df[j1_cols].apply(lambda col: (col == 'Yes').sum()).sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=j1_counts.head(5).values, y=j1_counts.head(5).index, palette='coolwarm')
plt.title('Top 5 Financial Problems')
plt.xlabel('Count')
plt.ylabel('Problem')
plt.tight_layout()
plt.show()

# --- Top Digital Access Barriers (K2__) ---
k2_cols = [col for col in df.columns if col.startswith('K2__')]
k2_counts = df[k2_cols].apply(lambda col: (col == 'Yes').sum()).sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=k2_counts.head(5).values, y=k2_counts.head(5).index, palette='viridis')
plt.title('Top 5 Digital Access Barriers')
plt.xlabel('Count')
```
